```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

dataset address: https://www.kaggle.com/piyushgoyal443/red-wine-dataset#wineQualityReds.csv

```python
data = pd.read_csv('wineQualityReds.csv', index_col=0)
data.head()
```

```python
data.dtypes
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
plt.figure(figsize=(14,8))
plt.scatter(np.log10(data['fixed.acidity']), np.log10(data['residual.sugar']), marker='o', 
            s=np.exp(data['alcohol']/3), c=data['quality'],
            cmap=plt.get_cmap("Reds", 5), alpha=0.5, label='alcohol')

cbar = plt.colorbar()
cbar.set_label('Wine Quality', fontsize=16)
plt.legend()
plt.xlabel('fixed.acidity')
plt.ylabel('residual.sugar')
plt.savefig('wine.png')
```

Figure 1. The scatter plot of wine dataset. The size stands for the alcohol percentage and the color map represents the wine quality. It shows that there does not exist a simple relation between acidity and residual sugar.
