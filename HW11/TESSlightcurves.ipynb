# **Computing Lomb-Scargle periodograms for each time series**

Using the package loaded above (LombScargle), compute the periodogram (either the power spectrum or the amplitude spectrum, as discussed in class) for each time series, and then plot each periodogram, making sure to properly label the axes and to identify each star. Make sure to choose the frequency range wisely.

For more documentation on the LombScargle package: https://docs.astropy.org/en/stable/api/astropy.timeseries.LombScargle.html#astropy.timeseries.LombScargle

For the normalization, make sure to use "psd" so that we can recover units.

```python
def plot_fq(fq, amp, **kwargs):
    
    plt.plot(fq, amp, **kwargs)
    plt.legend()
    plt.xlabel('frequency (d$^{-1}$)')
    plt.ylabel('amplitude (mmag)')

def plot_periodogram(t, y, dy=0.1, **kwargs):
    '''plot power spectrum'''
    fq, power = LombScargle(t, y, dy).autopower(normalization='psd', nyquist_factor=0.2)
    
    N = len(t)
    amp = 2 * (power/N) ** 0.5
    
    plot_fq(fq, amp, **kwargs)
    
    return fq, amp
```

```python
plt.figure(figsize=(16, 8))

f0, amp0 = plot_periodogram(star_0.bjd, star_0.mag, 
                 label='star 011400909', alpha=0.5 )
f1, amp1 = plot_periodogram(star_1.bjd, star_1.mag, 
                 label='star 146879345', alpha=0.5)
f2, amp2 = plot_periodogram(star_2.bjd, star_2.mag, 
                 label='star 277023834', alpha=0.5)
```

# **Phase folding**

For any star above exhibiting periodic variations, you will phase-fold their time series. This is accomplished by rescaling the exogenous variable, dividing it by the period (keep in mind that period = 1/frequency, and that if you have multiple periods, or harmonics, you should divide by the fundamental period). However, only the fractional part (e.g. np.modf(x)[0]) of the period must be considered (which means that the axis containing the exogenous variable will no longer be monotonically increasing; therefore, make sure to reorder the dataset appropriately).

Then, plot the phase-folded light curve(s) and identify them accordingly.

$\phi = \frac{t-t_0}{P}$

```python
period = 1.56
phi = (star_0.bjd - star_0.bjd[0]) / period

# select one period
phi_one = phi[phi <=1]
mag_one = star_0.mag[phi <=1]
plt.plot(phi_one, mag_one)
plt.xlabel('phase')
plt.ylabel('mmag')
```

Figure 8. The phase-folded light curve.

> From below figures we can see that binned light curve with 20 bins misses a lot of information, it is not appropriate. If we add to 50 bins, the light curve looks much better.

```python
from scipy.stats import binned_statistic
```

```python
## use 20 bins
nbins = 20
(values, edges, bin_number) = binned_statistic(phi_1, mag_1, statistic='mean', bins=nbins)

bined_phi = edges[0:len(edges)-1]
bined_mag = values
plt.plot(bined_phi, bined_mag, label='binned', linewidth=1.0)
plt.plot(phi_1, mag_1, label='non-binned', alpha=0.5)
plt.legend()
plt.xlabel('phase')
plt.ylabel('mmag')
```

Figure 9. The binned light curve using 20 bins. 

```python
## use more bins
nbins = 50
(values, edges, bin_number) = binned_statistic(phi_1, mag_1, statistic='mean', bins=nbins)

bined_phi = edges[0:len(edges)-1]
bined_mag = values
plt.plot(bined_phi, bined_mag, label='bined', linewidth=1.0)
plt.plot(phi_1, mag_1, label='non-bined', alpha=0.5)
plt.legend()
plt.xlabel('phase')
plt.ylabel('mmag')
```
